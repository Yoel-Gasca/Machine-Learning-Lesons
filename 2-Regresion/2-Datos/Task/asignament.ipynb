# Explorando visualizaciones
Se crearon algunas visualizaciones utilizando los datos de 'Pumpkin' en esta lección con matplotlib y seaborn en este cuaderno de muestra.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos desde el archivo CSV
df= pd.read_csv('../../Data/US-pumpkins.csv')
```

```python
df.head()
```

Se importan las librerías a utilizar y el archivo csv que se utiliza.

## Visualización 1: Histograma
Un histograma es útil para visualizar la distribución de una variable numérica. En este ejemplo, crearemos un histograma de una columna llamada 'Low Price' en el DataFrame donde estan los precios bajos:

```python
# Crear un histograma
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Low Price', bins=20, kde=True)
plt.xlabel('Low Price')
plt.ylabel('Frecuencia')
plt.title('Distribución de precios')
plt.show()
```

## Visualización 2: Gráfico de barras
Un gráfico de barras es útil para representar datos categóricos. En este ejemplo, se crea un gráfico de barras para mostrar la cantidad de los precios mas altos en la comlumna 'High Price':

```python
# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='High Price')
plt.xlabel('High Price')
plt.ylabel('Price')
plt.title('Distribución de Género')
plt.show()
```

Visualización 3: Diagrama de dispersión
Un diagrama de dispersión es útil para visualizar la relación entre dos variables numéricas. En este ejemplo, crearemos un diagrama de dispersión para mostrar la relación entre los precios altos y bajos en cada ciudad.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Low Price', y='High Price', hue='City Name')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.title('Relación entre Edad e Ingresos')
plt.show()
```

Estas son solo algunas visualizaciones básicas que se pueden crear con Matplotlib y Seaborn. Se pueden personalizar estas visualizaciones según las necesidades y explorar otras opciones de estas bibliotecas para crear visualizaciones más avanzadas.
