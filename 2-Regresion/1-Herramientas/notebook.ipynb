# Python y Scikit-learn para modelos de regresión

Este es un pequeño análisis de un conjunto de datos sobre la diabetes que está integrado con Scikit-learn con fines de aprendizaje. Imagínese que quisiera probar un tratamiento para pacientes diabéticos. Los modelos de Machine Learning, pueden ayudar a determinar que pacientes responderían mejor al tratamiento, en función de combinaciones de variables.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Se importan las bibliotecas matplottlib, numpy y también importando datasets, linear_model y model_selection de sklearn. model_selection se usa para dividir datos en conjuntos de entrenamiento y de prueba.

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

Se agregaron algunos comandos de impresión para mostrar la forma de la matriz de datos y su primer elemento:

## El conjunto de datos de diabetes

```python
X = X[:, np.newaxis, 2]
```

Se realiza una  seleccion una parte de este conjunto de datos para graficarlos colocándolos en una nueva matriz utilizando la función newaxis de numpy.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Se tiene los datos listos para graficarlos, puede ver si una máquina se puede ayudar a determinar una división lógica entre los númnero en este conjunto de datos. Para hacer esto, necesita dividir los datos (X) y el objetivo (y) en conjunto de datos de prueba y entrenamiento.

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Para entrenar el modelo! se carga el modelo de regresión lineal para entrenarlo con los datos de entrenamiento X e y usando model.fit():

```python
y_pred = model.predict(X_test)
```

Se creo una predicción usando datos de prueba, usando la función predict(). Esto se utilizará para trazar la línea entre los grupos de datos.

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

Se creo una gráfica de dispersión de todos los datos de prueba X e y, y se uso la predicción para dibujar una línea en el lugar más apropiado, entre las agrupaciones de datos del modelo.
